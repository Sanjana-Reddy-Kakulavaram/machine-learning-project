```
import numpy as np
import pandas as pd
```

```
data = pd.read_csv("amazon_baby.csv")
```

```
data.shape
```

```
data.head()
```

```
data.columns
```

```
data.dtypes
```

```
#checking for null values
data.apply(lambda x: sum(x.isnull()))
```

```
#drop the rows with no reviews
data.dropna(subset = ['name'], inplace = True)
data.dropna(subset = ['review'], inplace = True)

data.head()
```

```
#importing necessary packages
import os
import matplotlib as mpl
if os.environ.get('DISPLAY','') == '':
    print('no display found. Using non-interactive Agg backend')
    mpl.use('Agg')
import matplotlib.pyplot as plt
```

```
data.name.value_counts().head(10).plot(kind = 'bar')
plt.xticks(rotation=86)
plt.xlabel('UserID')
plt.ylabel('Number of Reviews')
plt.show()
```

```
#RATINGS COUNT
data['rating'].value_counts().plot(kind = 'bar')
plt.title('Number of Good Ratings vs Bad Ratings')
plt.xlabel('Rating Scales')
plt.xticks(rotation = 0)
plt.ylabel('Total ratings')
plt.show()
```

```
data['rating'].value_counts()
```

```
def remove(data):
    if data['rating'] == 3:
        return -1
    elif data['rating'] > 3:
        return 1
    return 0
```

```
data = data[data.rating != 3]
```

```
data['sentiment'] = np.where(data['rating'] > 2, 'positive', 'negative')
data['ratings'] = np.where(data['rating'] > 2, 1, 0)

data.head()
```

```
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```
stop_words = stopwords.words('english')
```

```
#removing stop words 
data['review'] = data['review'].apply(lambda x: ' '.join(word for word in x.split() if word not in stop_words))
```

```
#removing words that is non alpha
data['review'] = data['review'].apply(lambda x: ' '.join(word for word in x.split() if word.isalpha()))
```

```
#making all words to be lowercase
data['review'] = data['review'].apply(lambda x: ' '.join(word.lower() for word in x.split()))
```

```
data.sentiment.value_counts()
```

```
review_pos = data[data["sentiment"] == 'positive']
review_neg = data[data["sentiment"] == 'negative']
```

```
from wordcloud import WordCloud, STOPWORDS
```

```
text = review_pos["review"]
wordcloud = WordCloud(
    width = 800,
    height = 400,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(text))
fig = plt.figure(
    figsize = (20, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```
text = review_neg["review"]
wordcloud = WordCloud(
    width = 800,
    height = 400,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(text))
fig = plt.figure(
    figsize = (20, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```
from sklearn.model_selection import train_test_split
```

```
X = data['review']
y = data['ratings']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25)
```

```
print("X_train :",len(X_train))
print("X_test :", len(X_test))
print("y_train :", len(y_train))
print("y_test :", len(y_test))
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```
tfidf = TfidfVectorizer()
cls = LogisticRegression(max_iter=100000)
```

```
from sklearn.pipeline import Pipeline
model = Pipeline([('vectorizer', tfidf), ('classifier', cls)])
model.fit(X_train, y_train)
```

```
prediction = model.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix
print(confusion_matrix(prediction, y_test))
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction))
```

```
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test, prediction))
```

```
print("Training accuracy :", model.score(X_train, y_train))
print("Testing accuracy :", model.score(X_test, y_test))
```

```
from sklearn.metrics import accuracy_score, precision_score, recall_score

print("Accuracy :", accuracy_score(prediction, y_test))
print("Precision :", precision_score(prediction, y_test))
print("Recall :", recall_score(prediction, y_test))
```
